```python
from collections import Counter
from random import randint

runs = 100000

def get_winner(destinations):
    return Counter(destinations).most_common(1)[0][0]

def take_quiz():
    choices = []
    for question in answer_key:
        answer_index = randint(0,3)
        choices += answer_key[question][list(answer_key[question].keys())[answer_index]]
    return get_winner(choices)

winners = []
for _ in range(runs):
    winners.append(take_quiz())
total = len(winners)
for dest, count in Counter(winners).most_common():
    print(f"{dest}, {100*count/total:.2f}%")
```
